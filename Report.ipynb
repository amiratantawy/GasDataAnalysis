```python
transactions = pd.read_csv('Transactions.csv',low_memory=False)
customers = pd.read_csv('Customer.csv',low_memory=False)
transactions.head()
customers.head()
```

```python
data = transactions.merge(customers,left_on = 'TaqaID',right_on = 'TaqaID')
data.head()
```

```python
data.drop(['Custom','Average consumtion'], axis=1)
```

```python
data['date']= pd.to_datetime(data.date)
data['Date'] = data['date'].dt.strftime('%m/%d/%Y')
data['ConversionDate']= pd.to_datetime(data.ConversionDate)
data['ConversionDate2'] = data['ConversionDate'].dt.strftime('%m/%d/%Y')
data=data.drop(['date','ConversionDate'],axis=1)
```

```python
data['ConversionYear']= pd.DatetimeIndex(data['ConversionDate2']).year
data['CollectionYear']= pd.DatetimeIndex(data['Date']).year
data['ConversionMonth']= pd.DatetimeIndex(data['ConversionDate2']).month
data['CollectionMonth']= pd.DatetimeIndex(data['Date']).month
data=data.drop(['Custom'],axis=1)
data.head()
```

```python
AVG_Consumption_Per_Customer    = data.groupby('TaqaID').agg({'Consumotion':'mean'})
Total_Consumption_Per_Customer  = data.groupby('TaqaID').agg({'Consumotion':'sum'})
CNT_AccountMonth_Per_Customer   = data.groupby('TaqaID').agg({'AccountMonth':'count'})
Total_AccountMonth_Per_Customer = data.groupby('TaqaID').agg({'AccountMonth':'sum'})
AverageAccountMonthPerCustomer  = Total_AccountMonth_Per_Customer/CNT_AccountMonth_Per_Customer
Total_Customers_Per_year  = data.groupby('ConversionYear')['TaqaID'].nunique()
```

```python
AVG_Consumption_Per_year    = data.groupby('CollectionYear').agg({'Consumotion':'mean'})
```

```python
import matplotlib.pyplot as plt
AVG_Consumption_Per_year.plot(kind="line",title="Yearly Consumption")

plt.show()
```

```python
Total_Customers_Per_year.plot(kind="bar",title="Customers Count by year")
```

```python
Consumption_Per_Customer    = data.groupby('CustomerType').agg({'Consumotion':'sum'})
Consumption_Per_Customer.plot(kind='pie',title='consumption per customer type',subplots=True)
```

```python
Total_Customers_Per_metertype  = data.groupby('MeterType')['TaqaID'].nunique().sort_values()
Total_Customers_Per_metertype.plot(kind='bar', title='Total customers by meter type')
```

```python
Consumption_Per_metertype   = data.groupby('MeterType').agg({'Consumotion':'sum'}).sort_values(by='Consumotion')
Consumption_Per_metertype.plot(kind='barh',title='consumption per MeterType',subplots=True)
```

```python
city= pd.read_csv('city.csv')
city['Code']=city['Code'].astype(str)

data['TaqaID']=data['TaqaID'].astype(str)
data['city_id']= data['TaqaID'].str[:3]
city.head(1)
```

```python
data=data.merge(city,left_on='city_id',right_on='Code')
data.head(1)
```

```python
Consumption_Per_City   = data.groupby('City').agg({'Consumotion':'mean'}).sort_values(by='Consumotion')
Consumption_Per_City.plot(kind='bar',title='Average Consumption by city')
```

```python
Accountmonth_Per_City   = data.groupby('City').agg({'AccountMonth':'mean'}).sort_values(by='AccountMonth')
Accountmonth_Per_City.plot(kind='bar',title='Average AccountMonth by city')
```
